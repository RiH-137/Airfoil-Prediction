Since there is  no null values, therefore no need of data cleaning, eda or feature engineering. So we directly move onto the training part.

###     1. train test

```python
X= df.iloc[::-1]
y= df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test=train_test_split(X,y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
lr=regressor.fit(X_train, y_train)
lr
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
arr=scaler.fit_transform(X)
arr
```

```python
df1=pd.DataFrame(arr)
df1
```

```python
#prediction
test=lr.predict([[1.083165,1.490443,-0.373740,0.723345,3.172773,-2.991714]])
test
```

this is the case of overfitting as the accuracy is more than 100%

```python
#accurace
lr.score(X_test, y_test)
```

very poor accuracy of about 2%

```python
##using lasso
from sklearn.linear_model import Ridge, Lasso, RidgeCV, LassoCV, ElasticNet, ElasticNetCV, LinearRegression
```

```python
lassocv=LassoCV(cv=10, max_iter=20000000)
lassocv.fit(X_train, y_train)
```

```python
lassocv.score(X_test, y_test)
```

very poor accuracy

```python
ridgecv=RidgeCV(alphas=(0.1, 1.0, 10.0), cv=10)     #giving random 3 size to alphas
ridgecv.fit(X_train, y_train)
ridgecv.score(X_test, y_test)
```

very poor accuracy

```python
import pickle
pickle.dump(regressor, open('model.pkl', 'wb'))
```

```python
pickled_model=pickle.load(open('model.pkl','rb'))
#batch input
pickled_model.predict(X_test)
```

```python
pickled_model.score(X_test, y_test)
```
